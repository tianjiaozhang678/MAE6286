```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
pyplot.rcParams['font.family'] = 'serif'
pyplot.rcParams['font.size'] = 16
```

```python
def sokudo(rho):
    V = Vmax * (1 - rho/rho_max)
    return V
```

```python
nx = 51
L = 11.0
dx = L / (nx - 1)
dt = 0.001
c = 1.0

x = numpy.linspace(0.0, L, num=nx)
```

```python
rho0 = numpy.ones(nx)*10
rho0[10:20] = 50
```

```python
pyplot.figure(figsize=(4.0, 4.0))
pyplot.title('Initial conditions')
pyplot.xlabel('x')
pyplot.ylabel('rho')
pyplot.grid()
pyplot.plot(x, rho0, color='C0', linestyle='--', linewidth=2)
pyplot.xlim(0.0, L)
pyplot.ylim(0.0, 60.0)
```

```python
Vmax = 80
rho_max = 250
nt = 50
rho = rho0.copy()
for n in range(1, nt):
    rho[1:] = rho[1:] - Vmax * c * dt / dx * ((rho[1:] - rho[:-1]) - 1/rho_max * (rho[1:]**2 - rho[:-1]**2))
```

```python
pyplot.figure(figsize=(4.0, 4.0))
pyplot.xlabel('x')
pyplot.ylabel('rho')
pyplot.grid()
pyplot.plot(x, rho0, label='Initial',
            color='C0', linestyle='--', linewidth=2)
pyplot.plot(x, rho, label='nt = {}'.format(nt),
            color='C1', linestyle='-', linewidth=2)
pyplot.legend()
pyplot.xlim(0.0, L)
pyplot.ylim(0.0, 60.0)
```

```python
sum = 0.0
for i in range(nx - 1):
    sum += sokudo(rho[i])
    
AV = sum / (nx-1) / 3.6
print(AV)
```

```python
nt = 100
rho = rho0.copy()
for n in range(1, nt):
    rho[1:] = rho[1:] - Vmax * c * dt / dx * ((rho[1:] - rho[:-1]) - 1/rho_max * (rho[1:]**2 - rho[:-1]**2))

    
pyplot.figure(figsize=(9.0, 4.0))
pyplot.xlabel('x')
pyplot.ylabel('rho')
pyplot.grid()
pyplot.plot(x, rho0, label='Initial',
            color='C0', linestyle='--', linewidth=2)
pyplot.plot(x, rho, label='nt = {}'.format(nt),
            color='C1', linestyle='-', linewidth=2)
pyplot.legend()
pyplot.xlim(0.0, L + 2)
pyplot.ylim(0.0, 60.0)
```
